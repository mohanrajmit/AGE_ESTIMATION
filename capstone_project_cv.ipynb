```python
# importing the zipfile module
from zipfile import ZipFile
import numpy as np
import pandas as pd
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

###CNN Dependencies
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPool2D
from keras.layers import Dropout
from keras.layers import Flatten
from keras.layers import Dense
import cv2
from matplotlib import pyplot as plt
  

from keras.preprocessing.image import ImageDataGenerator
```

```python
#IMAGE_SIZE = 224 # Image size of resize when applying transforms.
BATCH_SIZE = 32
#NUM_WORKERS = 4 # Number of parallel processes for data preparation.
ROOT_PATH = '/content/stanford-car-dataset-by-classes-folder-224/stanford-car-dataset-by-classes-folder-224/'
#TEST_ANNOTATIONS_PATH = ROOT_PATH+'anno_test.csv'
#TRAIN_ANNOTATIONS_PATH = ROOT_PATH+'anno_train.csv'
#NAMES_PATH= ROOT_PATH+'names.csv'
```

```python
from imutils import paths

DATA_DIR_TRAIN = ROOT_PATH+'car_data/train/'
train_classes = list(paths.list_images(DATA_DIR_TRAIN))

DATA_DIR_TEST = ROOT_PATH+'car_data/test/'
test_classes = list(paths.list_images(DATA_DIR_TEST))

```

```python
train_classes
```

```python
train_images = []
train_labels = []
import cv2

for image_path in train_classes:

  image = cv2.imread(image_path)
  print(image.shape)
  image = cv2.resize(image,(128,128))
  train_images.append(image)
  label = image_path.split("/")[-2]
  train_labels.append(label)
```

```python
import numpy as np

train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
train_images.shape
```

```python
test_images = []
test_labels = []
import cv2

for test_image_path in test_classes:

  image_test = cv2.imread(test_image_path)
  #print(image.shape)
  image_test = cv2.resize(image_test,(128,128))
  test_images.append(image_test)
  test_label = test_image_path.split("/")[-2]
  test_labels.append(test_label)
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
test_images.shape
```

```python
no_classes = len(np.unique(train_labels))
```

```python
no_classes
```

```python
from sklearn.preprocessing import LabelEncoder
# encode the labels using LabelEncoder
le = LabelEncoder()
train_labels = le.fit_transform(train_labels)
from keras.utils import np_utils
y_train = np_utils.to_categorical(train_labels, no_classes)
```

```python
test_labels = le.fit_transform(test_labels)
y_test = np_utils.to_categorical(test_labels, no_classes)
```

```python
train_labels[0],y_train[0]
```

```python
cnn = Sequential()
#Adding 1st Convolution and Pooling Layer
cnn.add(Conv2D(32,kernel_size=(3,3),input_shape=(128,128,3),activation='relu',padding='same'))
cnn.add(Conv2D(32,kernel_size=(3,3),input_shape=(128,128,3),activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
#cnn.add(Dropout(0.2))
#Adding 2nd Convolution and Pooling Layer
cnn.add(Conv2D(64,kernel_size=(3,3),activation='relu',padding='same'))
cnn.add(Conv2D(64,kernel_size=(3,3),activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
#cnn.add(Dropout(0.2))
#Adding 3rd Convolution and Pooling Layer
cnn.add(Conv2D(128,kernel_size=(3,3),activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
#cnn.add(Dropout(0.2))
#Adding 4th Convolution and Pooling Layer
cnn.add(Conv2D(256,kernel_size=(3,3),activation='relu'))
cnn.add(MaxPool2D(pool_size=(2,2)))
#Flatten
cnn.add(Flatten())
cnn.add(Dense(units=256,activation='relu'))
cnn.add(Dropout(0.2))
cnn.add(Dense(units=no_classes,activation='softmax'))

```

```python
cnn.summary()
```

```python
train_images = train_images.astype('float32')

train_images /= 255
test_images = test_images.astype('float32')
test_images /= 255
```

```python
cnn.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = cnn.fit(train_images,y_train,epochs=10,batch_size=32,validation_data=(test_images,y_test))
```
